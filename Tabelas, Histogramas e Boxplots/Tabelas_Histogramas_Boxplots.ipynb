### Visualizando dados em Python
#### Tabelas, Histogramas e Boxplots

Quando trabalhar em um novo dataset, uma das coisas mais úteis a se fazer é começar a visualizar os dados. Usando tabelas, histogramas, boxplots e outras ferramentas visuais, conseguimos ter uma melhor ideia sobre o que os dados estão tentando nos contar, e conseguimos adquirir insights sobre os dados que talvez não seriam descobertos de outra maneira. 

Hoje, vamos aprender alguns conceitos básicos de visualização em Python, aprendendo a explorar dados a partir de uma perspectiva gráfica.


```
#Impordando os pacotes necessários
import seaborn as sns #Para plotagem
import matplotlib.pyplot as plt #Para mostrar os plots

#Carregando o Dataset, nesse caso o dataset será carregado a partir do 
#GitHub do criador do SeaBorn: https://github.com/mwaskom/seaborn-data

tips_data = sns.load_dataset("tips")
```

Isso nos mostrará qual o tipo de dados que estamos lidando, e servirá de referência para os outros gráficos que estamos prestes a fazer.

**Descrevendo Dados**

Um sumário estatístico, que geralmente inclui medidas como a média, valor mínimo e máximo dos dados, valores dos quartis e desvio padrão, podem ser úteis para saber a grandeza de algumas das variáveis e quais são as variáveis mais importantes.

```
#Imprimindo o sumário estatístico para variáveis quantitativas
tips_data.describe()
```

**Criando um Histograma**

Depois de criarmos uma noção sobre o que se tratam os dados, é uma boa ideia descobrir como os dados estão distribuídos. 

```
#Criando um histograma do valor total da conta
sns.displot(tips_data["total_bill"], kde = False)
plt.title("Valor total da Conta")
plt.show()
```

```
#Criando um histograma do total de gorjetas
sns.displot(tips_data["tip"], kde = False)
plt.title("Valor total de Gorjetas")
plt.show()
```

```
#Criando um histograma do total de gorjetas e total da conta
sns.histplot(tips_data["total_bill"], kde= False)
sns.histplot(tips_data["tip"], kde=False)
plt.title("Histograma de Total da Conta e Total de Gorjetas")
plt.show()
```

**Criando um Boxplot**

Boxplots não nos mostram o formato da distribuição, mas podem nos dar uma melhor ideia sobre o centro e a propagação dos dados, juntamente com possíveis outliers. Boxplots e Histogramas são geralmente complementares um ao outro e podem ajudar a conseguir mais informações sobre os dados.

```
#Criando um boxplot com os valores totais das contas
sns.boxplot(data = tips_data["total_bill"])
plt.title("Box plot dos valores totais das contas")
plt.show()
```

```
#Criando um boxplot com os valores totais das gorjetas
sns.boxplot(data = tips_data["tip"])
plt.title("Box plot das gorjetas")
plt.show()
```

```
#Criando os boxplots lado a lado
sns.boxplot(data = tips_data["total_bill"])
plt.title("Boxplot com os valores totais das contas")
plt.show()
sns.boxplot(data = tips_data["tip"])
plt.title("Boxplot com os valores totais das gorjetas")
plt.show()
```

**Criando Histogramas e Boxplots a partir de grupos**

Enquanto a análise de uma única variável é interessante, geralmente é útil analisar a variação de uma variável em resposta à outra. Usando gráficos, podemos ver se existe diferença na gorjeta dada por fumantes vs. não fumantes, se as gorjetas variam de acordo com o horário do dia, ou podemos então explorar tendências no conjunto de dados. 

```
#Criando um boxplot comparando a gorjeta dada por fumantes vs não fumantes
sns.boxplot(x = tips_data["tip"], y = tips_data["smoker"])
plt.show()
```

```
#Criando um boxplot e histogramas de gorjetas por horário do dia
sns.boxplot(x = tips_data["tip"], y = tips_data["time"])

g = sns.FacetGrid(tips_data, row = "time")
g = g.map(plt.hist, "tip")
plt.show()
```

```
#Criando boxplot e histogramas de gorjetas por dia 
sns.boxplot(x = tips_data["tip"], y = tips_data["day"])

g = sns.FacetGrid(tips_data, row = "day")
g = g.map(plt.hist, "tip")
plt.show()
```
